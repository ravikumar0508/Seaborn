```python
#Regression plot
import seaborn as sns
from seaborn import load_dataset
```

```python
tips =load_dataset("tips")
tips
```

```python
tips.head()
```

```python
#Regplot
sns.regplot(x='total_bill',y ="tip",data =tips)
```

```python
#lmplot
sns.lmplot(x='total_bill',y ="tip",hue ="smoker",markers=["o","x"],palette ="Set1",data =tips)
```

```python
#Data truncations
sns.lmplot(x='total_bill',y ="tip",hue ="smoker",markers=["o","x"],palette ="Set1",col="time",row="sex",data =tips)
```

```python
sns.lmplot(x='total_bill',y ="tip",hue ="smoker",markers=["o","x"],palette ="Set1",col="time",row="size",data =tips)
```

```python
sns.lmplot(x='total_bill',y ="tip",hue ="smoker",markers=["o","x"],palette ="Set1",data =tips)
```
